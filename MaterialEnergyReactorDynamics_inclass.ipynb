```python
import numpy as np
from scipy.integrate import solve_ivp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
```

```python
q = 100.  #L/min
cain = 1. #mol/L
Tin = 350. #K
V = 100. #L
rho = 1e3 #g/L
C = 0.239 #J/(g K)
hr = 5e4 #J/mol
ER = 8750. #K
k0 = 7.2e10 #1/min
UA = 5e4 #J/(min K)

Tc0 = 300. #K
ca0 = 0.5 #mol/L
T0 = 350. #K

Tc = 305. #K
tend = 10. #min
```

```python
def rhs(t, vec):
    ca, T = vec
    k = k0*np.exp(-ER/T)
    return [q/V*(cain - ca) - k*ca,
            q/V*(Tin-T) + hr*k*ca/(rho*C)- UA*(T-Tc)/(rho*V*C)]
```

```python
sol=solve_ivp(rhs, (0,tend), [ca0, T0], method='Radau', dense_output=True).sol
```

```python
tplot = np.linspace(0,tend,500)
ca, T = sol(tplot)
```

```python
fig = make_subplots(cols=2,rows=1)
fig.add_scatter(x=tplot, y=ca, mode='lines',row=1,col=1)
fig.add_scatter(x=tplot, y=T, mode='lines',row=1,col=2)
```
