```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
from scipy.integrate import solve_ivp
```

```python
def rhs(t, v):
    h1, h2, h3 = v
    h3sp=4.
    kc=0.1
    qin=kc*(h3sp-h3)
    q12 = np.sign(h1-h2)*alpha1*((np.abs(h1-h2))**0.5)
    q23 = np.sign(h2-h3)*alpha2*((np.abs(h2-h3))**0.5)
    q3 = alpha3*(h3**0.5)
    return [-q12/A1,
            (qin+q12-q23)/A2,
            (q23-q3)/A3]
```

```python
A1, A2, A3 = 0.5, 1., 2.
h01, h02, h03 = 0.5, 1., 2.
alpha1, alpha2, alpha3 = 0.1, 0.1, 0.1
tend=200.
```

```python
sol = solve_ivp(rhs, (0,tend), (h01, h02, h03), method='Radau',
                dense_output=True).sol
```

```python
t_plot=np.linspace(0,tend,500)
h1,h2,h3=sol(t_plot)
fig = make_subplots()
fig.add_scatter(x=t_plot, y=h1, mode='lines',name='h1')
fig.add_scatter(x=t_plot, y=h2, mode='lines',name='h2')
fig.add_scatter(x=t_plot, y=h3, mode='lines',name='h3')
```
