```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
from scipy.integrate import solve_ivp
from scipy.optimize import root
```

```python
def rhs(t, vec, casp):
    ca = vec[0]
    err = casp - ca
    cain = cain0 + kc*err
    return [q*(cain - ca)/V - 2*k*ca**2]
```

```python
k = 1e-5    # L/(mol s)
V = 20.     # L
q = 1.5     # L/s

tend = 100. # s
ca0 =   root(rhs, casp0).x[0] # mol/L
cain0 = 0.5 # mol/L
casp0 = 0.8 # mol/L
casp = 0.9  # mol/L
kc = 5.
```

```python
def wrap_rhs(vec):
    return rhs(0,vec, 0.8)
```

```python
root(wrap_rhs, [casp0]).x[0]
```

```python
sol=solve_ivp(rhs, (0,tend), [ca0], method='Radau', dense_output=True ).sol
```

```python
t_plot = np.linspace(0., tend, 500)
fig = make_subplots()
fig.add_scatter(x=t_plot, y=sol(t_plot)[0], mode='lines', name='ca')
```
