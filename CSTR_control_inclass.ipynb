```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default='plotly_dark'
from scipy.integrate import solve_ivp
from scipy.optimize import root
```

```python
def rhs(t, vec, casp):
    ca = vec[0]
    err = casp - ca
    cain = cain0 + kc*err
    return [q*(cain - ca)/V - 2*k*ca**2]
```

```python
k = 1e-5    # L/(mol s)
V = 20.     # L
q = 1.5     # L/s

tend = 100. # s
cain0 = 0.5 # mol/L
casp0 = 0.8 # mol/L
casp = 0.9  # mol/L
kc = 10.

def wrap_rhs_root(vec):
    return rhs(0,vec, casp0)

ca0 =   root(wrap_rhs_root, [casp0]).x[0]  # mol/L
```

```python
def wrap_rhs_solveivp(t, vec):
    return rhs(t,vec,casp)
```

```python
sol=solve_ivp(wrap_rhs_solveivp, (0,tend), [ca0], method='Radau', dense_output=True ).sol
```

```python
t_plot = np.linspace(0., tend, 500)
fig = make_subplots()
fig.add_scatter(x=t_plot, y=sol(t_plot)[0], mode='lines', name='ca')
```

```python
def rhs_PI(t, vec, casp):
    ca, errint = vec
    err = casp - ca
    cain = cain0 + kc*(err + errint/taui)
    return [q*(cain - ca)/V - 2*k*ca**2,
            err]
```

```python
k = 1e-5    # L/(mol s)
V = 20.     # L
q = 1.5     # L/s

tend = 100. # s
cain0 = 0.5 # mol/L
casp0 = 0.8 # mol/L
casp = 0.9  # mol/L
kc = 1.
taui = 10.

def wrap_rhs_PI_root(vec):
    return rhs_PI(0,vec, casp0)

ca0,errint0 =   root(wrap_rhs_PI_root, [casp0, 0.]).x  # mol/L
```

```python
def wrap_rhs_PI_solveivp(t, vec):
    return rhs_PI(t, vec, casp)
```

```python
sol = solve_ivp(wrap_rhs_PI_solveivp, [0,tend], [ca0, errint0], method='Radau',
          dense_output=True).sol
t_plot = np.linspace(0., tend, 500)
```

```python
ca, errint = sol(t_plot)
```

```python


fig2 = make_subplots()
fig2.add_scatter(x=t_plot, y=ca, mode='lines', name='ca')
```
