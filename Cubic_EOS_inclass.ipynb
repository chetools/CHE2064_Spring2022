```python
def cubic_abes(T, eos='SRK'):
    Tr = T/p.Tc
    e = {'vdw':0., 'SRK':0.}[eos]
    s = {'vdw':0., 'SRK':1.}[eos]
    omega = {'vdw': 1./8., 'SRK':0.08664}[eos]
    psi = {'vdw':27./64., 'SRK':0.42748}[eos]
    alpha = {'vdw': 1.,
             'SRK': (1. + (0.48 + 1.574 * p.w - 0.176*(p.w)**2) * 
                     (1. - Tr**0.5))**2}[eos]

    a = psi * alpha * (R**2) * (p.Tc)**2 / p.Pc
    b = omega * R * p.Tc/p.Pc
    return a, b, e, s
```

```python
def cubic_P(v, T, eos='SRK'):
    a, b, e, s = cubic_abes(T, eos)
    return R*T/(v-b) - a/((v+e*b)*(v+s*b))
```

```python
b = 0.08664 * R * p.Tc/p.Pc
v= np.logspace( np.log10(1.01*b), -2, 500)
fig = make_subplots()

for T in range(275, 800, 25):
    P=cubic_P(v, T)
    fig.add_scatter(x=v, y=P,mode='lines', name=f'T={T} K')
fig.update_xaxes(type='log')
fig.update_yaxes(range=[-200e6, 200e6])
```
